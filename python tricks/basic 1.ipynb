```python
# --- Case Modifications ---

# 1. capitalize() [00:00:32]
# Capitalizes the first character of the string.
text = "hello"
print(text.capitalize())  # Hello

# 2. casefold() [00:00:57]
# Returns a caseless version of the string for comparisons.
text1 = "Mario"
text2 = "MARIO"
print(text1.casefold() == text2.casefold())  # True

# 26. lower() [00:13:06]
# Converts the entire string to lowercase.
text = "UPPERCASE"
print(text.lower())  # uppercase

# 44. swapcase() [00:21:41]
# Swaps uppercase characters to lowercase, and vice-versa.
text = "Luigi has pasta"
print(text.swapcase())  # lUIGI HAS PASTA

# 45. title() [00:21:55]
# Converts the first character of each word to uppercase.
text = "this is a title"
print(text.title())  # This Is A Title

# 46. upper() [00:22:17]
# Converts the entire string to uppercase.
text = "hello there"
print(text.upper())  # HELLO THERE


# --- Searching & Counting ---

# 4. count() [00:02:07]
# Counts occurrences of a substring.
text = "abc_abc_abc_abc"
print(text.count("ab"))  # 4

# 8. find() [00:04:02]
# Returns the lowest index of the substring (-1 if not found).
text = "remember to comment and subscribe"
print(text.find("subscribe"))  # 24

# 11. index() [00:05:49]
# Similar to find(), but throws an error if not found.
text = "astronauts recently discovered a banana on the moon?"
print(text.index("banana"))

# 34. rfind() [00:16:52]
# Returns the highest index of the substring starting from the right.
text = "a some text.a"
print(text.rfind("a"))

# 35. rindex() [00:17:26]
# Similar to rfind(), but throws an error if not found.
text = "a some text.a"
print(text.rindex("a"))


# --- Formatting & Alignment ---

# 3. center() [00:01:37]
# Centers a string within a given amount of space.
text = "text"
print(text.center(20, "_"))

# 9. format() [00:04:52]
# Formats string using keywords or positional arguments.
text = "{subject} is doing {action}"
print(text.format(subject="cat", action="meow"))  # cat is doing meow

# 10. format_map() [00:05:21]
# Formats string using a dictionary.
coords = {"x": 10, "y": -5}
text = "Coordinates are {x} and {y}"
print(text.format_map(coords))

# 25. ljust() [00:12:38]
# Left-justifies the string within a given length.
text = "text"
print(text.ljust(20, "_"))

# 36. rjust() [00:17:51]
# Right-justifies the string within a given length.
text = "text"
print(text.rjust(20, "_"))

# 47. zfill() [00:22:30]
# Pads the string with zeros on the left until it reaches a given length.
text = "text"
print(text.zfill(10))  # 000000text


# --- Splitting & Joining ---

# 24. join() [00:11:52]
# Joins an array of strings using the given string as a separator.
arr = ["text1", "text2", "text3"]
print("-".join(arr))  # text1-text2-text3

# 30. partition() [00:15:00]
# Splits string into three parts at the first occurrence of the separator.
text = "a+b=c"
print(text.partition("="))  # ('a+b', '=', 'c')

# 37. rpartition() [00:18:03]
# Similar to partition(), but starts from the right.
text = "text=text2=text3"
print(text.rpartition("=")) 

# 38. rsplit() [00:18:37]
# Splits the string into a list from the right side.
text = "this is some special stuff"
print(text.rsplit(maxsplit=2))

# 39. split() [00:19:26]
# Splits the string into a list from the left side.
text = "this is some special stuff"
print(text.split(maxsplit=2))

# 41. splitlines() [00:20:23]
# Splits the string at line breaks (\n).
text = "remember\nto comment"
print(text.splitlines())


# --- Stripping & Replacing ---

# 27. lstrip() [00:13:21]
# Removes leading characters (spaces by default).
text = "   text"
print(text.lstrip())

# 31. removeprefix() [00:15:31]
# Removes a specific prefix from the string.
text = "wasapp"
print(text.removeprefix("was"))  # app

# 32. removesuffix() [00:15:45]
# Removes a specific suffix from the string.
text = "Mr everyone"
print(text.removesuffix("one"))  # Mr every

# 33. replace() [00:16:04]
# Replaces occurrences of a substring.
text = "remember to comment"
print(text.replace("comment", "subscribe", 1))

# 40. rstrip() [00:20:03]
# Removes trailing characters.
text = "his name is Mario Mario"
print(text.rstrip(" Mario"))

# 43. strip() [00:21:21]
# Removes both leading and trailing characters.
text = "Luigi pasta Luigi"
print(text.strip("Luigi"))


# --- Encoding & Translation ---

# 5. encode() [00:02:27]
# Encodes string into bytes (e.g. utf-8).
text = "Elon Musk"
print(text.encode(encoding="utf-8", errors="strict"))

# 7. expandtabs() [00:03:39]
# Expands \t tabs to a specified number of spaces.
text = "hello\tworld"
print(text.expandtabs(20))

# 28. maketrans() & 29. translate() [00:13:49]
# Creates a translation table and translates the string.
table = str.maketrans({"B": "🥓"})
text = "BABY"
print(text.translate(table))  # 🥓A🥓Y


# --- Boolean Checks ---

# 6. endswith() [00:03:01]
print("Apple".endswith(("e", "a")))  # True

# 12. isalnum() [00:06:25]
print("HelloKitty123".isalnum())  # True (letters and numbers only)

# 13. isalpha() [00:06:54]
print("HelloKitty".isalpha())  # True (letters only)

# 14. isascii() [00:07:08]
print("Text".isascii())  # True (checks Ashi/ASCII characters)

# 15, 16, 17. isdecimal(), isdigit(), isnumeric() [00:07:27]
# Checks for numeric characteristics (decimal is strictest, numeric is broadest).
print("123".isdecimal())  # True
print("①②③".isdigit())  # True
print("一二三".isnumeric()) # True

# 18. isidentifier() [00:09:42]
print("test_sample".isidentifier())  # True (valid python variable name)

# 19. islower() [00:10:06]
print("abc".islower())  # True

# 20. isprintable() [00:10:27]
print("text\n".isprintable())  # False (due to \n)

# 21. isspace() [00:10:54]
print("   ".isspace())  # True

# 22. istitle() [00:11:09]
print("This Is A Title".istitle())  # True

# 23. isupper() [00:11:38]
print("BANANAS".isupper())  # True

# 42. startswith() [00:20:59]
print("Luigi".startswith("L"))  # True
```
