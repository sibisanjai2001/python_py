```python
#basic linked list node implementation

class first:
    def __init__(self,data):
        self.data=data
        self.next=None

a=first(10)
b=first(20)

a.next=b

print(id(b))
print(a.data,a.next)
print(a.data,a.next.data)
print("======================")
a.next.data=30
print(a.data,a.next.data)
```

```python
a=[45,75,85,45]
target=45+85


for i in range(len(a)):
    for j in range(i+1,len(a)):
        if a[i]+a[j]==target:
            print(i,j)
```
