```python
class sibi:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return self.name + " says hello!"
    

s=sibi("Sibi")
s2=sibi("kavin")

print(id(s))
print(id(s2))
print(id(s.name))
print(id(s2.name))

#    ❌ Misconception: You might think s.name "is part of" s, so they should share the same ID.
#   ✅ Reality: The instance s contains a reference to the string object s.name, but they are distinct objects in memory.
#s is an instance of the class sibi.
#s.name is an attribute of that instance—specifically, a string object.
#    s → is the instance object of class sibi.
#      s.name → is a separate string object ("Sibi") stored as an attribute inside the instance. 
""""
Summary:

    id(s) → identity of the object instance.
    id(s.name) → identity of the string object that is an attribute of s.
    They are different objects, so their IDs are supposed to be different. ✅
    """
```

```python
a = "hello"
b = "hellos"
print(id(a) == id(b))  # Often True (but not guaranteed for all strings)
```
