```python
# number of line = number rows
#----  |----> outer loops should run
for i in range(5): 
    for j in range (5):
        print(end="*")
    print() # outer loop for number of rows 
```

```python
for i in range(5): 
    for j in range (7):
        print(end="*")
    print() 
```

```python
n=int(input("Enter number of rows: "))
for i in range (n+1):
    for j in range (i):
        print("*", end="")
    print()
```

```python
for i in range (10,0,-1):
    for j in range (i):
        print(end="*")
    print()
```
