```python
import matplotlib as plt
import numpy as np
import pandas as pd
df = pd.DataFrame({
    "Date": pd.date_range(start="2023-01-01", periods=100, freq="D"),
    "Sales": np.random.randint(100, 1000, size=100),
    "Temperature": np.round(np.random.uniform(10, 35, size=100), 1),
    "Product_Category": np.random.choice(["Electronics", "Clothing", "Groceries"], size=100),
    "Customers": np.random.randint(20, 200, size=100)
    }).head(20)
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
# print(plt.__version__)
plt.plot(df['Sales'],df['Temperature'])
plt.show()
```

```python
x= pd.DataFrame({1,55,3,4,5})
y= pd.DataFrame({2,3,4,5,6})
# plt.plot(x,y,"s")
plt.plot(x,y,marker="o")
plt.title("Sales vs Temperature")
plt.xlabel("Sales")
plt.ylabel("Temperature")
plt.show()
```
